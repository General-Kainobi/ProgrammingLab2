### **1. Distribuzioni di probabilità**
Genera 1000 valori casuali da una distribuzione gamma con parametro di forma pari a 1.
Suggerimento: il parametro di forma viene passato come primo argomento quando si "congela" la distribuzione.

Traccia l’istogramma del campione e sovrapponi la PDF della distribuzione.

Stima il parametro di forma dal campione usando il metodo fit.

Extra:
Le distribuzioni hanno molti metodi utili. Esplorali usando il completamento automatico con il tasto TAB.

Traccia la funzione di distribuzione cumulativa (CDF).

Calcola la varianza.

```python
import pandas as pd
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
gen_frozen=sp.stats.gamma(a=1)
randgen=gen_frozen.rvs(1000)

plt.hist(randgen, density=True)
x=np.linspace(0, max(randgen), 100)
plt.plot(x,gen_frozen.pdf(x))
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Gamma Distributio')
plt.legend()

shape=sp.stats.gamma.fit(randgen)
print("shape = ", shape)

plt.figure(figsize=(10, 6))
plt.plot(x, gen_frozen.cdf(x), 'b-', lw=2, label='CDF')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.title('Gamma Distribution CDF')
plt.legend()

variance = gen_frozen.var()
print("var:", variance)
plt.show
```

### **2. Fitta i dati**
Prova a fittare i dati sottostante con le migliori curve, calcola il MAE e l'RMSE

```python
import numpy as np
temp_max = np.array([17, 19, 21, 28, 33, 38, 37, 37, 31, 23, 19, 18])
temp_min = np.array([-62, -59, -56, -46, -32, -18, -9, -13, -25, -46, -52, -58])
import matplotlib.pyplot as plt
months = np.arange(12)
plt.plot(months, temp_max, "ro")
plt.plot(months, temp_min, "bo")
plt.xlabel("Month")
plt.ylabel("Min and max temperature")
fit= np.polyfit(months, temp_max, deg=4)
fits=sp.interpolate.UnivariateSpline(months, temp_max, s=2)
y_smooth = fits(months)
plt.plot(y_smooth)
def calculate_mae(actual, predicted):
    return np.mean(np.abs(actual - predicted))

def calculate_rmse(actual, predicted):
    return np.sqrt(np.mean((actual - predicted) ** 2))

y_pred_spline = y_smooth 

mae_spline = calculate_mae(temp_max, y_pred_spline)
rmse_spline = calculate_rmse(temp_max, y_pred_spline)

print("Spline Curve MAE:", mae_spline)
print("Spline Curve RMSE:", rmse_spline)


fits_min = sp.interpolate.UnivariateSpline(months, temp_min, s=2)
y_smooth_min = fits_min(months)
plt.plot(y_smooth_min, label="Spline Fit (Min Temp)", color="red")


y_pred_spline_min = y_smooth_min
mae_spline_min = calculate_mae(temp_min, y_pred_spline_min)
rmse_spline_min = calculate_rmse(temp_min, y_pred_spline_min)

# Print the error metrics for temp_min
print("Spline Curve MAE (Min Temp):", mae_spline_min)
print("Spline Curve RMSE (Min Temp):", rmse_spline_min)

plt.legend()
plt.show()
```
